```python
# required library for data analysis
import pandas as pd
import numpy as np

# required librairies for data visualization
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context("talk")
```

#### Loading the dataset

```python
df = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/Python/02 Python Data Visualization/Data_Visualization_with_Python_s2-main/telecom_data.csv")

display(df.head(7))
```

```python
display(df.describe())
```

```python
df['Intl_Plan'].dtypes
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(y = "Total_charges", data=df, showfliers=False)
```

**Findings from Boxplot**
1. min
2. first quartile (25th percentile)
3. median (50th percentile)
4. third quartile (75th percentile)
5. max
6. range = max-min
7. IQR (inter-quartile range) = third quartile(70%) - first quartile(25%)

```python
df.columns
```

```python
num_cols = ['Account_Length', 'Vmail_Message', 'CustServ_Calls', 
            'Total_mins', 'Total_calls', 'Total_charges']
```

#### Separate the numerical columns

```python
num_cols = []

for col in df.columns:
    if df[col].dtypes != "O":
        num_cols.append(col)
         

print(num_cols)
```

```python
num_cols[5]
```

#### Create a for loop to show the boxplots of all numeric features

```python
for index in range(len(num_cols)):
    
     print("index = ", index, "position = ", index+1, "column name = ", num_cols[index])
    
```

```python
plt.figure(figsize=(20,8)) # width, height

for index in range(len(num_cols)):
    plt.subplot(2,3,index+1)
    sns.boxplot(y=num_cols[index], data=df, showfliers=False)   #y=num_cols[index]

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,8)) # width, height

for index in range(len(num_cols)):
    
    
    plt.subplot(2,3,index+1)
    sns.boxplot(x=num_cols[index], data=df, showfliers=False)   #x=num_cols[index]

plt.suptitle("Boxplot of the numeric columns in the telecom dataset")
plt.tight_layout()
plt.show()
```

#### Adding categorical features in the boxplot

```python
plt.figure(figsize=(20,8)) # width, height

for index in range(len(num_cols)):
    plt.subplot(2,3,index+1)
    sns.boxplot(x=num_cols[index], data=df, showfliers=False, y=df["Churn"])

plt.suptitle("Boxplot of the numeric columns in the telecom dataset")
plt.tight_layout()
plt.show()
```

**Drawback of box plot**
- could not find any information about Avg
- could not find any information about **Data Set**

```python
df["Churn"].value_counts()
```

#### Adding categorical features (hue) in the boxplot

```python
plt.figure(figsize=(25,12)) # width, height

for index in range(len(num_cols)):
    plt.subplot(2,3,index+1)
    sns.boxplot(y=df[num_cols[index]], x=df["Intl_Plan"], hue=df["Churn"], showfliers=False)

plt.suptitle("Boxplot of the numeric columns in the telecom dataset")
plt.tight_layout()
plt.show()
```

#### Comparing boxplots and violinplot

```python
plt.figure(figsize=(20,8)) # width, height

for index in range(len(num_cols)): 
    plt.subplot(2,3,index+1)
    sns.violinplot(x=num_cols[index], data=df, showfliers=False)

plt.suptitle("Boxplot of the numeric columns in the telecom dataset")
plt.tight_layout()
plt.show()
```

**white dot** mediun value

```python
plt.figure(figsize=(25,12)) # width, height

for index in range(len(num_cols)): 
    plt.subplot(2,3,index+1)
    sns.violinplot(x=df[num_cols[index]], y=df["Churn"], showfliers=False)

plt.suptitle("Boxplot of the numeric columns in the telecom dataset")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(25,12)) # width, height

for index in range(len(num_cols)): 
    plt.subplot(2,3,index+1)
    sns.violinplot(y=df[num_cols[index]], x=df["Intl_Plan"], hue=df["Churn"], showfliers=False, split=True)

plt.suptitle("Boxplot of the numeric columns in the telecom dataset")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(25,12)) # width, height

for index in range(len(num_cols)): 
    plt.subplot(2,3,index+1)
    sns.violinplot(x=df[num_cols[index]], y=df["Intl_Plan"], hue=df["Churn"], showfliers=False)

plt.suptitle("Boxplot of the numeric columns in the telecom dataset")
plt.tight_layout()
plt.show()
```
